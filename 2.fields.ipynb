### <b><u><i>3.creating multilevel validation


# 🔍 Multilevel Validation in Pydantic

Multilevel validation refers to how **Pydantic** 🔧 validates **nested or structured data** at multiple levels — such as:

- 🔢 Fields  
- 📚 Lists  
- 🗂️ Dictionaries  
- 🧩 Nested models

It doesn't just validate the **top-level types** — it **recursively dives** into the entire data structure to ensure everything matches the expected format. 🧠✅

---

## 🏥 Example: `Patient` Model

In the example below, the `Patient` model contains complex fields that illustrate **multilevel validation**:

### 🎯 Fields

```python
from typing import List, Dict
from pydantic import BaseModel

class Patient(BaseModel):
    allergies: List[str]
    contact_details: Dict[str, str]
```

### 🔍 Field Breakdown

- **`allergies: List[str]`**  
  📌 A list where **every item must be a string**.

- **`contact_details: Dict[str, str]`**  
  📌 A dictionary where **both keys and values must be strings**.

---

## 🧠 How It Works

Pydantic **recursively validates** all the values:

✅ Ensures every item in `List[str]` is a `str`  
✅ Ensures all keys and values in `Dict[str, str]` are `str`  
✅ Applies the same rules to **deeply nested models** and structures

---

## 🚀 Summary

Multilevel validation in Pydantic helps you:

- Catch deeply hidden errors 🐞
- Maintain data integrity 🔐
- Handle nested models like a pro 🧑‍💻

> 💡 **Pro Tip:** Combine multilevel validation with custom validators (`@validator`, `@root_validator`) for even more control!
