# <b>Nested Fields

```python
from pydantic import BaseModel

class Address(BaseModel):
  city:str
  state:str
  pin:str


class Patient(BaseModel):
  name:str
  age:int
  gender:str
  address:Address


address_dict={
    'city':'gurgaon',
    'state':'haryana',
    'pin':'12234'
}

address1=Address(**address_dict)

patient_dict={
    'name':'vikas',
    'gender':'male',
    'age':22,
    'address':address1
}

patient1=Patient(**patient_dict)

print(patient1)
print(patient1.name)
print(patient1.address.city)
print(patient1.address.pin)

```
