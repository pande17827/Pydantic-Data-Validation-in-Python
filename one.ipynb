```python
from pydantic import BaseModel
```

```python
class Patient(BaseModel):
  name:str
  age:int
```

```python
patient_info={'name':'nitish',
              'age':30}
```

```python
patient1=Patient(**patient_info) # unpacking the dictionary
```

```python
def insert_patient(patient:Patient):
  print(patient.name)
  print(patient.age)
  print('inserted')
```

```python
insert_patient(patient1)
```
